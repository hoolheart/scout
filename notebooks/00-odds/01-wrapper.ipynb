# 创建杯套图案


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d0 = 7.0
d1 = 8.0
h = 8.2
```

```python
R = d1 / (d1 - d0) * h
R
```

```python
theta = d1 * np.pi / R
theta / np.pi * 180.0
```

```python
theta0 = d0 * np.pi / (R - h)
theta0 / np.pi * 180.0
```

```python
angles = np.linspace(-theta / 2.0, theta / 2.0, 21, True)
angles / np.pi * 180.0
```

```python
upper_x = (R - h * 0.4) * np.sin(angles)
upper_y = (R - h * 0.4) * np.cos(angles)
lower_x = (R - h * 0.8) * np.sin(angles)
lower_y = (R - h * 0.8) * np.cos(angles)
left_x = [upper_x[0], lower_x[0]]
left_y = [upper_y[0], lower_y[0]]
right_x = [upper_x[-1], lower_x[-1]]
right_y = [upper_y[-1], lower_y[-1]]
```

```python
plt.figure()
plt.plot(upper_x, upper_y, '-k')
plt.plot(lower_x, lower_y, '-k')
plt.plot(left_x, left_y, '-k')
plt.plot(right_x, right_y, '-k')
ax = plt.gca()
ax.set_aspect('equal')
ax.xaxis.set_units('cm')
```
