# CAN Setting

Calculate propriate CAN setting for the given CAN bus.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print (f"Import numpy {np.__version__}")
print (f"Import matplotlib {matplotlib.__version__}")
print (f"Import pandas {pd.__version__}")
```

```python
requirements = [
    (16, 500),
    (24, 100),
    (24, 125),
    (24, 250),
    (24, 500),
    (24, 800),
    (24, 1000),
    (48, 100),
    (48, 125),
    (48, 250),
    (48, 500),
    (48, 800),
    (48, 1000),
    (120, 100),
    (120, 125),
    (120, 250),
    (120, 500),
    (120, 800),
    (120, 1000),
]
```

```python
for req in requirements:
    clk = req[0] * 1e3
    bd = req[1]
    presdiv = np.ceil(clk / bd / 16)
    total = int(clk / bd / presdiv)
    best_sample = 0.875
    if bd > 800:
        best_sample = 0.75
    elif bd > 500:
        best_sample = 0.8
    Pseg2 = int(np.round(total * (1 - best_sample)))
    if Pseg2 < 2:
        Pseg2 = 2
    left = total - Pseg2 - 1
    Progseg = int(np.round(left * 0.67))
    if Progseg > 8:
        Progseg = 8
    Pseg1 = left - Progseg
    print (f'{clk*1e-3}M, {bd}K, {presdiv}, {clk / bd / presdiv}, {total}, {Progseg}, {Pseg1}, {Pseg2}, '
           f'{((Progseg + Pseg1 + Pseg2 + 1.0) * presdiv * bd)}, '
           f'{(1.0 + Progseg + Pseg1) / (1.0 + Progseg + Pseg1 + Pseg2)}')
    # print('    {' f'FLEXCAN_BAUDCLK_HZ_{req[0]}M, FLEXCAN_BAUD_{bd}K, {int(presdiv)}, {Progseg}, {Pseg1}, {Pseg2}, 1' '},')
```
